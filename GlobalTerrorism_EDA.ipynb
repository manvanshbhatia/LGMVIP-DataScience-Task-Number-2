### Skewness

```python
#.skew() gives the value of skewness for each column
skewness = df.skew()
skewness
```

```python
#To get the skewed columns
skewed_columns = skewness[abs(skewness) > 0.5].index
skewed_columns
```

```python
#To get non-skewed columns
normal_columns = skewness[abs(skewness) < 0.5].index
normal_columns
```
